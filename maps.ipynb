```python
import pandas as pd
import folium

# Read UFO sighting data into a Pandas DataFrame
ufo_df = pd.read_csv('UFO_csv.csv', encoding='latin-1')

# Create a map centered around California
ca_map = folium.Map(location=[36.7783, -119.4179], zoom_start=6)

# Add markers for California UFO sightings
for index, row in ufo_df[ufo_df['State'] == 'CA'].iterrows():
    lat, lng = row['lat'], row['lng']
    date, time = row['Date'], row['Time']
    city, state = row['City'], row['State']
    shape, summary = row['Shape'], row['Summary']
    country = row['Country']

    # Create a popup message with sighting details
    popup_text = f"<b>{country} - {state}</b><br>Date: {date}<br>Time: {time}<br>City: {city}<br>Shape: {shape}<br>Summary: {summary}"

    # Add a marker for each sighting
    folium.Marker(location=[lat, lng], popup=popup_text, icon=folium.Icon(color='blue')).add_to(ca_map)

# Save the California map to an HTML file
ca_map.save('california_ufo_map.html')

# Create a map centered around New York
ny_map = folium.Map(location=[42.6526, -73.7562], zoom_start=7)

# Add markers for New York UFO sightings
for index, row in ufo_df[ufo_df['State'] == 'NY'].iterrows():
    lat, lng = row['lat'], row['lng']
    date, time = row['Date'], row['Time']
    city, state = row['City'], row['State']
    shape, summary = row['Shape'], row['Summary']
    country = row['Country']

    # Create a popup message with sighting details
    popup_text = f"<b>{country} - {state}</b><br>Date: {date}<br>Time: {time}<br>City: {city}<br>Shape: {shape}<br>Summary: {summary}"

    # Add a marker for each sighting
    folium.Marker(location=[lat, lng], popup=popup_text, icon=folium.Icon(color='blue')).add_to(ny_map)

# Save the New York map to an HTML file
ny_map.save('new_york_ufo_map.html')

# Create a map centered around Texas
tx_map = folium.Map(location=[31.9686, -99.9018], zoom_start=6)

# Add markers for Texas UFO sightings
for index, row in ufo_df[ufo_df['State'] == 'TX'].iterrows():
    lat, lng = row['lat'], row['lng']
    date, time = row['Date'], row['Time']
    city, state = row['City'], row['State']
    shape, summary = row['Shape'], row['Summary']
    country = row['Country']

    # Create a popup message with sighting details
    popup_text = f"<b>{country} - {state}</b><br>Date: {date}<br>Time: {time}<br>City: {city}<br>Shape: {shape}<br>Summary: {summary}"

    # Add a marker for each sighting
    folium.Marker(location=[lat, lng], popup=popup_text, icon=folium.Icon(color='blue')).add_to(tx_map)

# Save the Texas map to an HTML file
tx_map.save('texas_ufo_map.html')
```
