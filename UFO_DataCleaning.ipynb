```python
import csv
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
#Importing and Displaying the dataframe

ufo_df = pd.read_csv('data/uncleaned_data/UFO_CSV_Uncleaned.csv', encoding='latin-1')
ufo_df
```

```python
#First, making a series
ids = pd.Series(np.arange(len(ufo_df)))


#Second, insert series into the dataframe
ufo_df['Report_ID'] = ids
ufo_df
```

```python
#Reorder columns so the ID comes first

ufo_df = ufo_df[['Report_ID', 'Date', 'Time', 'Country', 'City', 'State', 'Shape', 'Summary', 'lat', 'lng']]
ufo_df
```

```python
#Checking for any null values in the dataframe

ufo_df.isnull().values.any()
```

```python
#Getting a sum of how many null values exist in the dataframe and which columns contain them

ufo_df.isnull().sum()
```

```python
#Creating another dataframe to show the rows of each null value for the 'Time' column (since 'Time' is the only column that\
#contains null values)

#Stackoverflow post assisted with the development of this syntax below. The post can be located at: https://stackoverflow.com/questions/67372679/how-to-check-isnull-for-the-list

bool_series = pd.isnull(ufo_df["Time"]) 
ufo_df[bool_series]
```

```python
#Dropping all 9 null values in the dataset

ufo_df.dropna(inplace=True)
ufo_df
```

```python
#Checking if any values are still null

ufo_df.isnull().values.any()
```

```python
#Exporting dataframe locally for SQL upload

ufo_df.to_csv('data/cleaned_data/UFO_CSV_Cleaned.csv', index=False)
```
